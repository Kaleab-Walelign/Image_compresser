```python
A = imread("/home/kali/Downloads/wexor-tmg-L-2p8fapOA8-unsplash.jpg")
X = np.mean(A,-1)
```

```python
#img = plt.imshow(256-X)
#img.set_cmap('gray')
#plt.axis('off')
#plt.show()
```

```python
U,S,VT = np.linalg.svd(X,full_matrices=False)
S = np.diag(S)

j = 0
for r in (5,35,100):
    Xapprox = U[:,:r] @ S[0:r,:r] @VT[:r,:]
    plt.figure(j+1)
    j+=1
    img = plt.imshow(256-Xapprox)
    img.set_cmap('gray')
    plt.axis('off')
    plt.title('Image_Compress_by = ' + str(r))
    plt.show()
```
